```python
import ipywidgets as widgets

#import plotly.graph_objs as go
import plotly.graph_objects as go

import pandas as pd
import numpy as np


def cubes(size, pos_x, pos_y, pos_z, color):
    # create points
    x, y, z = np.meshgrid(
        np.linspace(pos_x-size/2, pos_x+size/2, 2), 
        np.linspace(pos_y-size/2, pos_y+size/2, 2), 
        np.linspace(pos_z-size/2, pos_z+size/2, 2),
    )
    x = x.flatten()
    y = y.flatten()
    z = z.flatten()
    
    return go.Mesh3d(x=x, y=y, z=z, alphahull=1, flatshading=True, color=color)#,  contour_show=False)
                     #lighting={'diffuse': 0.1, 'specular': 2.0, 'roughness': 0.5})


# Read data from a csv
#z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')
z_data = pd.read_csv('mt_bruno_elevation.csv')

# Your colorscale options are (but not necessarily limited to):
# Greys,YlGnBu,Greens,YlOrRd,Bluered,RdBu,Reds,Blues,Picnic,Rainbow,
# Portland,Jet,Hot,Blackbody,Earth,Electric,Viridis,Cividis.

# colorscale – Sets the colorscale. The colorscale must be an array
#  containing arrays mapping a normalized value to an rgb, rgba, hex,
#  hsl, hsv, or named color string. At minimum, a mapping for the
#  lowest (0) and highest (1) values are required. For example,
#  [[0, 'rgb(0,0,255)'], [1, 'rgb(255,0,0)']]. To control the bounds 
#  of the colorscale in color space, use cmin and cmax. 
#  Alternatively, colorscale may be a palette name string of the
#  following list: Blackbody,Bluered,Blues,C ividis,Earth,Electric,
#  Greens,Greys,Hot,Jet,Picnic,Portl and,Rainbow,RdBu,Reds,Viridis,YlGnBu,YlOrRd.

fig = go.Figure()#FigureWidget()

fig.update_layout(title=dict(text='Mt Bruno Elevation 3'), autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
# fig.update_layout(
#    # showlegend=False,
#     xaxis=dict(visible=False),
#     yaxis=dict(visible=False),
# )
fig.update_layout(scene = dict(xaxis = dict( title=dict(text=''),showgrid = False,showticklabels = False,showbackground=False,showspikes=False),
                                   yaxis = dict( title=dict(text=''),showgrid = False,showticklabels = False,showbackground=False,showspikes=False),
                                   zaxis = dict( title=dict(text=''),showgrid = False,showticklabels = False,showbackground=False,showspikes=False)
             ))

# set edge length of cubes
size = 2

# add outer cube
fig.add_trace(cubes(size,20,20,100, 'rgba(255,100,0,1)'))

fig.add_trace(go.Mesh3d(
        x=[0, 1, 2, 0],
        y=[0, 0, 1, 2],
        z=[100, 120, 100, 110],
        color='green',
        # colorbar=dict(title=dict(text='z')),
        # colorscale=[[0, 'gold'],
        #             [0.5, 'mediumturquoise'],
        #             [1, 'magenta']],
        # Intensity of each vertex, which will be interpolated and color-coded
        # intensity=[0, 0.33, 0.66, 1],
        # i, j and k give the vertices of triangles
        # here we represent the 4 triangles of the tetrahedron surface
        i=[0, 0, 0, 1],
        j=[1, 2, 3, 2],
        k=[2, 3, 1, 3],
        #hoverinfo="none"
        # name='y',
        # showscale=True
    ))



fig.update_traces(hoverinfo="none")#,contours_x_highlight=False,contours_y_highlight=False,contours_z_highlight=False)
#fig.update_traces(contours=dict(x=dict(project=dict(x=False))))
#fig.update_traces(contours_x_highlight=False)
#fig.update_traces(contours={"contours_x_highlight":False, "contours_y_highlight":False, "contours_z_highlight":False})
#fig.update_scenes(xaxis_showspikes=False, yaxis_showspikes=False, zaxis_showspikes=False)
# add inner center cube


fig.add_trace(go.Surface(z=z_data.values, colorscale ='Earth',hoverinfo="x+y+z"))


play = widgets.Play(
    value=50,
    min=0,
    max=100,
    step=1,
    interval=500,
    description="Press play",
    disabled=False
)
slider = widgets.IntSlider()
widgets.jslink((play, 'value'), (slider, 'value'))
container = widgets.HBox([play, slider])
widgets.VBox([container,
              fig])
```
